```python
import pandas as pd
import numpy as np

# Create the dataset
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

print("Original DataFrame:")
print(df)

# ---- Handle Missing Values ----

# Impute 'age' with median
age_median = df['age'].median()
df['age'].fillna(age_median, inplace=True)

# Impute 'city' with mode
city_mode = df['city'].mode()[0]
df['city'].fillna(city_mode, inplace=True)

# ---- Remove Duplicate Rows ----
df = df.drop_duplicates()

print("\nCleaned DataFrame:")
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['city'])

# Label Encoding
label_encoder = LabelEncoder()
df_encoded['gender'] = label_encoder.fit_transform(df_encoded['gender'])

print("\nEncoded DataFrame:")
print(df_encoded)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Select numerical columns
numerical_cols = ['age', 'annual_income']

# ---- Min-Max Scaling ----
minmax_scaler = MinMaxScaler()

df_minmax = df_encoded.copy()
df_minmax[numerical_cols] = minmax_scaler.fit_transform(
    df_minmax[numerical_cols]
)

print("\nMin-Max Scaled Data:")
print(df_minmax)

# ---- Standardisation ----
standard_scaler = StandardScaler()

df_standard = df_encoded.copy()
df_standard[numerical_cols] = standard_scaler.fit_transform(
    df_standard[numerical_cols]
)

print("\nStandardised Data:")
print(df_standard)
```
